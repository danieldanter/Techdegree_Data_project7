# 1 Combine all of the CSV files into a single dataset using Pandas. 

```python
# read all scvs
y2010 = pd.read_csv("2010.csv", index_col="Unnamed: 0") 
y2011  = pd.read_csv("2011.csv", index_col="Unnamed: 0")
y2012  = pd.read_csv("2012.csv", index_col="Unnamed: 0")
y2013 = pd.read_csv("2013.csv", index_col="Unnamed: 0") 
y2014  = pd.read_csv("2014.csv", index_col="Unnamed: 0")
y2015  = pd.read_csv("2015.csv", index_col="Unnamed: 0")
y2016  = pd.read_csv("2016.csv", index_col="Unnamed: 0")
y2017  = pd.read_csv("2017.csv", index_col="Unnamed: 0")
y2018  = pd.read_csv("2018.csv", index_col="Unnamed: 0")
y2019  = pd.read_csv("2019.csv", index_col="Unnamed: 0")
```

# 10 Which genre has the highest energy score?

```python
# format the dataframe -> so we can group by genre and sum energy values
energy_score = all_top_songs.drop(['year', 'bpm','dnce','dB','live','val', 'dur', 'acous', 'spch','pop'], axis=1)
energy_score= energy_score.groupby(['top genre'])["nrgy"].sum().sort_values( ascending=False)
energy_score.head(15)
```

```python
# create bar chart for genre by sum of engegy score
energy_score.plot(kind='bar')
 
# labels for x & y axis
plt.xlabel('Genre')
plt.ylabel('Sum energy Score')
 
# title of plot
plt.title('genre by sum of engegy score')
```

### the answer is dance pop with an energy sum value of 23814
